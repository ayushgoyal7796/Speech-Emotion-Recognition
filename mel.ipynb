```python
import librosa
import librosa.display
import os
import warnings
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
angry = []
for root, dirs, files in os.walk(r'speech/angry/'):
    for file in files:
        if file.endswith('.wav'):
            angry.append(file)
disgust = []
for root, dirs, files in os.walk(r'speech/disgust/'):
    for file in files:
        if file.endswith('.wav'):
            disgust.append(file)
fearful = []
for root, dirs, files in os.walk(r'speech/fearful/'):
    for file in files:
        if file.endswith('.wav'):
            fearful.append(file)
happy = []
for root, dirs, files in os.walk(r'speech/happy/'):
    for file in files:
        if file.endswith('.wav'):
            happy.append(file)
neutral = []
for root, dirs, files in os.walk(r'speech/neutral/'):
    for file in files:
        if file.endswith('.wav'):
            neutral.append(file)
sad = []
for root, dirs, files in os.walk(r'speech/sad/'):
    for file in files:
        if file.endswith('.wav'):
            sad.append(file)
surprised = []
for root, dirs, files in os.walk(r'speech/surprised/'):
    for file in files:
        if file.endswith('.wav'):
            surprised.append(file)
```

```python
for i in angry:
    y, sr = librosa.load('speech/angry/'+i, duration=2.97)
    ps = librosa.feature.melspectrogram(y=y, sr=sr)
    librosa.display.specshow(librosa.power_to_db(ps, ref=np.max))
    plt.savefig('mel/angry/'+i.split('.')[0]+'.png', bbox_inches='tight', transparent=True, pad_inches=-0.05)
```
